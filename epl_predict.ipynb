```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df0=pd.read_csv('2021-2022.csv')
df1=pd.read_csv('2020-2021.csv')
df2=pd.read_csv('2019-20.csv')
df3=pd.read_csv('2018-19.csv')
```

```python
# To merge the cells 
df_merge=pd.concat([df0,df1,df2,df3],axis=0)
df_merge=df_merge.reset_index(drop=True)
```

```python
# Converting the data column to from object to dateTime 
df_merge['Date'] = pd.to_datetime(df_merge['Date'], errors='coerce')
```

```python
# changing some columns into interger, so it can be used in the machine learning
df_merge['Home_code']=df_merge['HomeTeam'].astype("category").cat.codes
df_merge['Away_code']=df_merge['AwayTeam'].astype("category").cat.codes
df_merge['Half_timeResult']=df_merge['HTR'].astype("category").cat.codes
df_merge['Target']=df_merge['FTR'].astype("category").cat.codes
df_merge['Day_code']=df_merge['Date'].dt.dayofweek
df_merge.head()
```

```python
# Either of the two code will remove the null value 
df_merge=df_merge.dropna(axis=0,subset=['Time'])
# df_merge=df_merge['Time'].replace(np.nan,0)   #this code will replace the null value with 0
```

```python
# To the only the hour of the column, some do well at a certain our 
df_merge['Hour']=df_merge['Time'].str.replace(":.+",'',regex=True).astype("int")  #without the dot ":+", the time will show hour and minutes
```

```python
df_merge.dtypes.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=50,min_samples_split=10,random_state=1)
```

```python
train=df_merge[df_merge['Date']<'2022-01-01']
test=df_merge[df_merge['Date']>='2022-01-01']   #tryin to ensure that I didn't use future data in training
```

```python
# trying to ensure I get as much columns as possible for me to get a better result 
columns=['Hour','HS','AS','FTHG','FTAG','HST','AST','HTHG','HTAG','HF','AF','HC','AC','HY','AY','HR','AR','Home_code','Away_code','Day_code','Half_timeResult']
```

```python
rf.fit(train[columns],train['Target'])
```

```python
predict=rf.predict(test[columns])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accu=accuracy_score(test['Target'],predict)
accu
```

```python
# I want to combine the actual result and what I predicted
combine=pd.DataFrame(dict(actual=test['Target'],prediction=predict))
```

```python
# To cross check our result with I predicted
pd.crosstab(index=combine['actual'],columns=combine['prediction'])
```
